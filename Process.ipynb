```python
import re
import pandas as pd

df = pd.read_csv("emails.csv")
```

```python
if "text" not in df.columns or "spam" not in df.columns:
    raise ValueError("There should be text' ve 'spam' columns in csv file.")
```

```python
df["label"] = df["spam"].astype(int)

df["text"] = df["text"].fillna("").astype(str)

clean_texts = []
```

```python
for text in df["text"]:

    text = text.lower()

    text = re.sub(r"http\S+|www\.\S+", " <URL> ", text)

    text = re.sub(r"\S+@\S+", " <EMAIL> ", text)

    text = re.sub(r"\d+", " <NUM> ", text)

    text = re.sub(r"[^a-z0-9\s<>]", " ", text)

    text = re.sub(r"\s+", " ", text).strip()

    clean_texts.append(text)
```

```python
df["text_clean"] = clean_texts

df = df[df["text_clean"].str.len() > 0].copy()

out_df = df[["text_clean", "label"]].reset_index(drop=True)
```

```python
out_df.to_csv("processed_emails.csv", index=False)

print("✅")
```
